```python
import os

# 1. Define the Kaggle path
# Replace 'chicago-crime' with whatever you named your dataset when you uploaded it
DATA_PATH = '/kaggle/input/datasets/nirmeetp/chicago-crime-dataset/Crimes_-_2001_to_Present.csv' 

# 2. Verify the file exists (This helps debug path errors)
if os.path.exists(DATA_PATH):
    print(f"File found at: {DATA_PATH}")
else:
    # This prints all available files so you can find the correct path if the above fails
    for dirname, _, filenames in os.walk('/kaggle/input'):
        for filename in filenames:
            print(os.path.join(dirname, filename))
```

```python

allCrimes = dd.read_csv(
    DATA_PATH,
    dtype={
        'X Coordinate': 'float64',
        'Y Coordinate': 'float64',
        'Ward': 'float64',
        'Community Area': 'float64',
        'District': 'float64',
        'FBI Code':'object',
        'IUCR':'object',
        'ID':'object'
    },
    assume_missing=True
)

print(allCrimes.dtypes)
print("Partitions:", allCrimes.npartitions)
```

```python
# Removing duplicates
allCrimes = allCrimes.map_partitions(lambda df: df.drop_duplicates(subset="ID"))
print("Dedup done — no cross-partition shuffle needed")
```

```python
# Removing NaN values
#(removing entire row if a NaN value is present)
allCrimes = allCrimes.dropna()
print("Records after dropping NaN:", len(allCrimes))
```

```python
# Removing invalid coordinates
#(remove where coordinates are 0, they are clearly wring/missing)


allCrimes = allCrimes[
    (allCrimes['X Coordinate'] != 0) &
    (allCrimes['Y Coordinate'] != 0) &
    (allCrimes['Longitude'] != 0) &
    (allCrimes['Latitude'] != 0)
]
print("Records after removing bad locations:", len(allCrimes))
```

```python
# convert date column from string to object
allCrimes['Date'] = dd.to_datetime(allCrimes['Date'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
# extracting time features from date
allCrimes['Hour']      = allCrimes['Date'].dt.hour
allCrimes['Month']     = allCrimes['Date'].dt.month
allCrimes['DayOfWeek'] = allCrimes['Date'].dt.dayofweek
allCrimes['Year']      = allCrimes['Date'].dt.year
```

```python
# keep top 25 crime locations
# Compute value counts (this triggers execution)
loc_counts = allCrimes['Location Description'].value_counts().compute()
mostFrequentLocations = list(loc_counts[:25].index)

# Replace rare locations with 'OTHER'
allCrimes['Location Description'] = allCrimes['Location Description'].where(
    allCrimes['Location Description'].isin(mostFrequentLocations),
    other='OTHER'
)

# Same for Description column
desc_counts = allCrimes['Description'].value_counts().compute()
mostFrequentDescription = list(desc_counts[:25].index)

allCrimes['Description'] = allCrimes['Description'].where(
    allCrimes['Description'].isin(mostFrequentDescription),
    other='OTHER'
)
```

```python
# convert to categorical for increasing speed
allCrimes['Primary Type']         = allCrimes['Primary Type'].astype('category')
allCrimes['Location Description'] = allCrimes['Location Description'].astype('category')
allCrimes['Description']          = allCrimes['Description'].astype('category')
```

```python
# drop unnecessary columns
allCrimes = allCrimes.drop(
    columns=['Case Number', 'IUCR', 'Updated On', 'FBI Code', 'Beat', 'Ward'],
    errors='ignore'
)
```

```python
# THIS is where all the above lazy steps actually execute (this is lazy evaluation of Dask)
allCrimes = allCrimes.compute()

print("Final clean dataset shape:", allCrimes.shape)
allCrimes.head()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, hour, to_timestamp, dayofweek

# Initialize Spark with a bit more memory overhead if possible
spark = SparkSession.builder \
    .appName("ChicagoCrime") \
    .config("spark.driver.memory", "8g") \
    .getOrCreate()

# 1. Load Data (Double-check this path matches your sidebar!)
path = "/kaggle/input/datasets/nirmeetp/chicago-crime-dataset" 
df_spark = spark.read.csv(path, header=True, inferSchema=True)

# 2. Process Timestamps
df_spark = df_spark.withColumn("Timestamp", to_timestamp(col("Date"), "MM/dd/yyyy hh:mm:ss a"))
df_spark = df_spark.withColumn("Hour", hour(col("Timestamp")))
df_spark = df_spark.withColumn("DayOfWeek", dayofweek(col("Timestamp")))

# 3. Clean and REPARTITION
# We include 'Primary Type' for analysis and 'Arrest' for modeling
df_clean = df_spark.select(
    "Primary Type", "Latitude", "Longitude", 
    "Hour", "DayOfWeek", "District", "Arrest"
).dropna()

# FIX: Break data into 200 chunks to prevent Disk Spilling warnings
df_clean = df_clean.repartition(200)

print(f"Data ready and partitioned. Total rows: {df_clean.count()}")
```

```python
from pyspark.sql.functions import year, col
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark import StorageLevel

df_clean = df_clean.filter((year(col("Timestamp")) > 2001) & (year(col("Timestamp")) < 2026))

print(f"Post-filter row count: {df_clean.count()}")

# 2. Vectorize (Latitude, Longitude, Hour)
assembler = VectorAssembler(inputCols=["Latitude", "Longitude", "Hour"], outputCol="raw_features")
v_data = assembler.transform(df_clean)

# 3. Scale the features
scaler = StandardScaler(inputCol="raw_features", outputCol="scaledFeatures", withStd=True, withMean=False)
scaler_model = scaler.fit(v_data)
final_data = scaler_model.transform(v_data)

# 4. FIX: Persist using Serialized format
final_data.persist(StorageLevel.MEMORY_AND_DISK)

print("Features scaled and vacuum-packed into memory.")
```

```python
from pyspark.ml.clustering import BisectingKMeans

# Train the model on the full 7.8M rows
# K=30 creates the crime zones we need
bkm = BisectingKMeans().setK(30).setSeed(42).setFeaturesCol("scaledFeatures")
cluster_model = bkm.fit(final_data)

# Assign Cluster IDs (This creates the 'prediction' column)
clustered_data = cluster_model.transform(final_data)

# Cache the final result for the next steps 
clustered_data.cache()

print("Clustering complete. The 'Big Blob' has been broken into 30 distinct zones.")
clustered_data.select("Primary Type", "prediction").show(5)
```

```python
# cnt of how many crimes fell into each cluster
cluster_counts = clustered_data.groupBy("prediction").count().orderBy("count", ascending=False)

print("--- Crime Distribution Across Clusters ---")
cluster_counts.show(n=30)
```

## MODELLING

```python
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.sql.functions import col

# --- STEP 1: RENAME & LABELING ---
# Rename the cluster ID column so it doesn't clash with the model's prediction
indexed_data = clustered_data.withColumnRenamed("prediction", "Hotspot_ID")

# Convert 'Primary Type' text to numbers
indexer = StringIndexer(inputCol="Primary Type", outputCol="typeIndex", handleInvalid="skip")
indexed_data = indexer.fit(indexed_data).transform(indexed_data)

# Convert 'Arrest' boolean to numeric 
indexed_data = indexed_data.withColumn("label", col("Arrest").cast("double"))

# --- STEP 2: FEATURE ASSEMBLY ---
# We include the Hotspot_ID and typeIndex—this is where the accuracy comes from!
feature_cols = ["typeIndex", "Hotspot_ID", "Hour", "DayOfWeek", "Latitude", "Longitude"]
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
final_model_data = assembler.transform(indexed_data)

# --- STEP 3: DATA SPLIT ---
train_data, test_data = final_model_data.randomSplit([0.8, 0.2], seed=42)

# --- STEP 4: UPGRADED MODEL CONFIGURATION ---
# numTrees=100 provides more stable predictions
# maxDepth=10 allows the model to see specific patterns (e.g. 'Narcotics' in 'Hotspot 5')
rf = RandomForestClassifier(
    labelCol="label", 
    featuresCol="features", 
    numTrees=100,      # IMPROVEMENT: Increase from 20 to 100
    maxDepth=10,       # IMPROVEMENT: Deeper trees catch more arrests
    maxBins=40, 
    predictionCol="arrest_prediction",
    seed=42
)

print("Training the High-Accuracy Model. This uses full CPU power...")
rf_model = rf.fit(train_data)

# --- STEP 5: PREDICTION & EVALUATION ---
predictions = rf_model.transform(test_data)

evaluator = BinaryClassificationEvaluator(labelCol="label", metricName="areaUnderROC")
auc = evaluator.evaluate(predictions)

print("\n" + "="*30)
print(f"IMPROVED MODELING COMPLETE")
print(f"Area Under ROC (AUC): {auc:.4f}")
print("="*30 + "\n")

# Show the 'Arrest-Heavy' crimes first to see the 1.0s in action
predictions.filter(col("arrest_prediction") == 1.0) \
    .select("Primary Type", "Hotspot_ID", "label", "arrest_prediction") \
    .show(20)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# 1. Precision: When the model predicts an arrest, how often is it right?
evaluator_p = MulticlassClassificationEvaluator(labelCol="label", predictionCol="arrest_prediction", metricName="weightedPrecision")
print(f"Precision: {evaluator_p.evaluate(predictions):.4f}")

# 2. Recall: Of all actual arrests, how many did the model find?
evaluator_r = MulticlassClassificationEvaluator(labelCol="label", predictionCol="arrest_prediction", metricName="weightedRecall")
print(f"Recall: {evaluator_r.evaluate(predictions):.4f}")

# 3. F1-Score: The balance between Precision and Recall
evaluator_f1 = MulticlassClassificationEvaluator(labelCol="label", predictionCol="arrest_prediction", metricName="f1")
print(f"F1 Score: {evaluator_f1.evaluate(predictions):.4f}")
```
